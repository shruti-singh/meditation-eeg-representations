```python
file_names = ["RESULTS_T2USER_MED_BINARYPRED_meditation.csv",
             "RESULTS_T3USER_MED_QUADPRED_meditation.csv",
             "RESULTS_T4USER_MED_BINARYPRED_meditation_CTRvsHT.csv",
             "RESULTS_T4USER_MED_BINARYPRED_meditation_CTRvsSNY.csv",
             "RESULTS_T4USER_MED_BINARYPRED_meditation_CTRvsVIP.csv",
             "RESULTS_T5USER_MED_TransferPRED_meditation_CTRvsHT.csv",
             "RESULTS_T5USER_MED_TransferPRED_meditation_CTRvsSNY.csv",
             "RESULTS_T5USER_MED_TransferPRED_meditation_CTRvsVIP.csv"]

if dtype == 'mind-wandering':
    new_file_names = []
    for f in file_names:
        new_file_names.append(f.replace("meditation", "mind-wandering"))
    file_names = new_file_names
```

```python
final_results = pd.DataFrame(columns=['Classifier', 'Ctr / Exp', 'Ctr / H / S / V', 
                                      'Ctr / H', 'Ctr / S',  'Ctr / V', 
                                      'Ctr / S-V', 'Ctr / H-V', 'Ctr / H-S'])
```

```python
file_path = "/home/singh_shruti/workspace/rep_learning/meditation-eeg-representations/iconip_sub/tradML/{}/".format(dtype)
all_models = list(pd.read_csv(file_path + file_names[0]).iloc[:, 0])
```

```python
final_results.head()
```

```python
# Read the first file
df = pd.read_csv(file_path+file_names[0])
df.head(3)
```

```python
for rowidx, row in enumerate(df.iterrows()):
    all_run_acc = [row[1]['{}_Test_Acc'.format(test_acc_idx)] for test_acc_idx in range(1, 6)]
    final_results.loc[rowidx] = row[1][0]
    # IN the line below, the 1 is the column index
    final_results.loc[rowidx][1] = "$" + str(round(np.mean(all_run_acc), 3)) +' \pm ' + str(round(np.std(all_run_acc), 3))+ "$"
final_results.head()
```

```python
for colidx in range(1, 5):
    df = pd.read_csv(file_path + file_names[colidx])
    for rowidx, row in enumerate(df.iterrows()):
        all_run_acc = [row[1]['{}_Test_Acc'.format(test_acc_idx)] for test_acc_idx in range(1, 6)]
        final_results.loc[rowidx][colidx+1] = "$" + str(round(np.mean(all_run_acc), 3)) +' \pm ' + str(round(np.std(all_run_acc), 3))+ "$"
```

```python
final_results.head()
```

```python
for colidx in range(5, 8):
    df = pd.read_csv(file_path + file_names[colidx])
    for rowidx, row in enumerate(df.iterrows()):
        all_run_acc = [row[1]['{}_Test_Transfer_Acc'.format(test_acc_idx)] for test_acc_idx in range(1, 6)]
        final_results.loc[rowidx][colidx+1] = "$" + str(round(np.mean(all_run_acc), 3)) +' \pm ' + str(round(np.std(all_run_acc), 3)) + "$"
```
